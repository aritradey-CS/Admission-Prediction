

1. **Data Loading**: We start by loading a dataset from a CSV file. This dataset contains information about students, including features like CGPA, GRE Score, TOEFL Score, and the target variable, which is "Chance of Admit" to a university.

2. **Feature and Target Definition**: We specify which columns in the dataset are considered features (X1, X2, X3) and the target variable we want to predict, which is "Chance of Admit."

3. **Data Splitting**: We split the dataset into two parts: a training set and a testing set. The training set is used to train our models, while the testing set is used to evaluate their performance. In this case, we allocate 20% of the data to the testing set.

4. **Model Creation and Training**:
   - We create two regression models: a Decision Tree Regressor and a Random Forest Regressor.
   - The Decision Tree model is a tree-based model that makes predictions based on a hierarchical structure of decisions.
   - The Random Forest model is an ensemble of multiple Decision Trees, which collectively make predictions.
   - We train both models using the training data, which involves learning the relationships between the features (CGPA, GRE Score, TOEFL Score) and the target variable ("Chance of Admit").

5. **Model Prediction**: After training, we use both models to make predictions on the testing set. These predictions are stored in the variables `y_pred_decision_tree` and `y_pred_random_forest`.

6. **Model Evaluation**: We calculate two evaluation metrics for both models: Mean Squared Error (MSE) and R-squared (R). These metrics help us understand how well our models are performing.
   - MSE measures the average squared difference between the actual and predicted values. Lower MSE indicates better performance.
   - R measures the proportion of the variance in the target variable that is predictable from the features. A higher R indicates a better fit.

7. **Printing Results**: We print the evaluation results for both models, displaying the MSE and R for each. This provides insights into how well the models are doing.

8. **Visualization**: Finally, we create a scatter plot to visualize the predicted values versus the actual values for both the Decision Tree and Random Forest models. This plot helps us assess how closely the models' predictions align with the actual data points.

In summary, this code loads a dataset, trains two regression models (Decision Tree and Random Forest), evaluates their performance using metrics, and visually compares their predictions to the actual values. This allows us to understand how well these models predict a student's chance of admission to a university based on their CGPA, GRE Score, and TOEFL Score.

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load your dataset (replace with your dataset path)
df = pd.read_csv('custom_data.csv')

# Define the main features (X1, X2, X3)
features = ['CGPA', 'GRE Score', 'TOEFL Score']

# Define the target variable
target = 'Chance of Admit'

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Decision Tree Regression
decision_tree_model = DecisionTreeRegressor(random_state=42)
decision_tree_model.fit(X_train, y_train)

# Random Forest Regression
random_forest_model = RandomForestRegressor(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_decision_tree = decision_tree_model.predict(X_test)
y_pred_random_forest = random_forest_model.predict(X_test)

# Visualize the predicted vs. actual values for Decision Tree
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_decision_tree, alpha=0.5)
plt.xlabel("Actual Chance of Admit")
plt.ylabel("Predicted Chance of Admit")
plt.title("Decision Tree Model")

# Visualize the predicted vs. actual values for Random Forest
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_random_forest, alpha=0.5)
plt.xlabel("Actual Chance of Admit")
plt.ylabel("Predicted Chance of Admit")
plt.title("Random Forest Model")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Load your dataset (replace with your dataset path)
df = pd.read_csv('custom_data.csv')

# Define the main features (X1, X2, X3)
features = ['CGPA', 'GRE Score', 'TOEFL Score']

# Define the university groups
universities = ['U1', 'U2', 'U3', 'U4', 'U5']

# Create dictionaries to store k-NN models and performance metrics
university_models = {}
university_mse = {}

# Iterate through each university group
for university in universities:
    # Filter students for the specific university group
    university_students = df[df['University Rating'] == university]

    # Check if there are enough samples for the university
    if len(university_students) < 2:
        print(f"University {university} does not have enough samples for modeling.")
        continue

    # Select the features (X) and target (Y) variables
    X = university_students[features]
    Y = university_students['Chance of Admit']

    # Split the data into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    # Create a k-NN regression model
    model = KNeighborsRegressor(n_neighbors=5)

    # Fit the model on the training data
    model.fit(X_train, Y_train)

    # Make predictions on the test data
    Y_pred = model.predict(X_test)

    # Calculate Mean Squared Error (MSE) as a performance metric
    mse = mean_squared_error(Y_test, Y_pred)

    # Store the k-NN model and MSE
    university_models[university] = model
    university_mse[university] = mse

# Print the Mean Squared Error for each university
for university, mse in university_mse.items():
    print(f"University: {university}")
    print(f"Mean Squared Error (MSE): {mse}")
    print("\n")
```

```python
import pandas as pd
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Load your dataset (replace with your dataset path)
df = pd.read_csv('custom_data.csv')

# Define the main features (X1, X2, X3)
features = ['CGPA', 'GRE Score', 'TOEFL Score']

# Define the university groups
universities = ['U1', 'U2', 'U3', 'U4', 'U5']

# Create dictionaries to store Ridge and Lasso models, as well as performance metrics
ridge_models = {}
lasso_models = {}
ridge_mse = {}
lasso_mse = {}

# Iterate through each university group
for university in universities:
    # Filter students for the specific university group
    university_students = df[df['University Rating'] == university]

    # Check if there are enough samples for the university
    if len(university_students) < 2:
        print(f"University {university} does not have enough samples for modeling.")
        continue

    # Select the features (X) and target (Y) variables
    X = university_students[features]
    Y = university_students['Chance of Admit']

    # Split the data into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    # Create Ridge and Lasso regression models
    ridge_model = Ridge(alpha=1.0)
    lasso_model = Lasso(alpha=1.0)

    # Fit the Ridge model on the training data
    ridge_model.fit(X_train, Y_train)

    # Fit the Lasso model on the training data
    lasso_model.fit(X_train, Y_train)

    # Make predictions on the test data for Ridge and Lasso
    ridge_Y_pred = ridge_model.predict(X_test)
    lasso_Y_pred = lasso_model.predict(X_test)

    # Calculate Mean Squared Error (MSE) for Ridge and Lasso
    ridge_mse_value = mean_squared_error(Y_test, ridge_Y_pred)
    lasso_mse_value = mean_squared_error(Y_test, lasso_Y_pred)

    # Store the Ridge and Lasso models and MSE values
    ridge_models[university] = ridge_model
    lasso_models[university] = lasso_model
    ridge_mse[university] = ridge_mse_value
    lasso_mse[university] = lasso_mse_value

# Print the Mean Squared Error for each university (Ridge and Lasso)
for university, mse in ridge_mse.items():
    print(f"University: {university}")
    print(f"Ridge Regression Mean Squared Error (MSE): {mse}")
    print("\n")

for university, mse in lasso_mse.items():
    print(f"University: {university}")
    print(f"Lasso Regression Mean Squared Error (MSE): {mse}")
    print("\n")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load your dataset (replace with your dataset path)
df = pd.read_csv('custom_data.csv')

# Define the features and target variable
features = ['CGPA', 'GRE Score', 'TOEFL Score']
target = 'Chance of Admit'

# Create a dictionary to store the evaluation results for each subgroup
evaluation_results = {}

# Iterate through each subgroup
for rating, subgroup in grouped_Univ:
    # Split the data into features (X) and target variable (Y)
    X = subgroup[features]
    Y = subgroup[target]

    # Split the data into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    # Create a Linear Regression model
    model = LinearRegression()

    # Train the model
    model.fit(X_train, Y_train)

    # Make predictions on the test set
    Y_pred = model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(Y_test, Y_pred)
    r2 = r2_score(Y_test, Y_pred)

    # Store the evaluation results for this subgroup
    evaluation_results[f'University Rating {rating}'] = {
        'MSE': mse,
        'R-squared': r2
    }

# Create lists to store MSE and R-squared values
mse_values = [results['MSE'] for results in evaluation_results.values()]
r2_values = [results['R-squared'] for results in evaluation_results.values()]

# Create a bar chart to visualize MSE and R-squared for each category
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Bar chart for MSE
axs[0].bar(evaluation_results.keys(), mse_values, color='skyblue')
axs[0].set_title('Mean Squared Error (MSE) for Each University Rating')
axs[0].set_ylabel('MSE')
axs[0].set_xlabel('University Rating')
axs[0].grid(axis='y', linestyle='--', alpha=0.6)

# Bar chart for R-squared
axs[1].bar(evaluation_results.keys(), r2_values, color='salmon')
axs[1].set_title('R-squared for Each University Rating')
axs[1].set_ylabel('R-squared')
axs[1].set_xlabel('University Rating')
axs[1].grid(axis='y', linestyle='--', alpha=0.6)

# Adjust layout for better readability
plt.tight_layout()

# Show the bar charts
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load your dataset (replace with your dataset path)
df = pd.read_csv('custom_data.csv')

# Define the main features (X1, X2, X3)
features = ['CGPA', 'GRE Score', 'TOEFL Score']

# Define the target variable
target = 'Chance of Admit'

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Create a dictionary to store evaluation results for each University Rating category
evaluation_results = {}

# Iterate through each University Rating category
for rating in df['University Rating'].unique():
    # Filter the data for the specific University Rating category
    subset_X_train = X_train[df['University Rating'] == rating]
    subset_X_test = X_test[df['University Rating'] == rating]
    subset_y_train = y_train[df['University Rating'] == rating]
    subset_y_test = y_test[df['University Rating'] == rating]

    # Create and fit a Decision Tree regression model
    decision_tree_model = DecisionTreeRegressor(random_state=42)
    decision_tree_model.fit(subset_X_train, subset_y_train)

    # Make predictions on the test set
    y_pred_decision_tree = decision_tree_model.predict(subset_X_test)

    # Calculate Mean Squared Error (MSE)
    mse = mean_squared_error(subset_y_test, y_pred_decision_tree)

    # Calculate R-squared
    r2 = r2_score(subset_y_test, y_pred_decision_tree)

    # Store the evaluation results for this category
    evaluation_results[rating] = {'MSE': mse, 'R-squared': r2}

# Create lists to store MSE and R-squared values
mse_values = [results['MSE'] for results in evaluation_results.values()]
r2_values = [results['R-squared'] for results in evaluation_results.values()]

# Create a bar chart to visualize MSE and R-squared for each category
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Bar chart for MSE
axs[0].bar(evaluation_results.keys(), mse_values, color='skyblue')
axs[0].set_title('Mean Squared Error (MSE) for Each University Rating')
axs[0].set_ylabel('MSE')
axs[0].set_xlabel('University Rating')
axs[0].grid(axis='y', linestyle='--', alpha=0.6)

# Bar chart for R-squared
axs[1].bar(evaluation_results.keys(), r2_values, color='salmon')
axs[1].set_title('R-squared for Each University Rating')
axs[1].set_ylabel('R-squared')
axs[1].set_xlabel('University Rating')
axs[1].grid(axis='y', linestyle='--', alpha=0.6)

# Adjust layout for better readability
plt.tight_layout()

# Show the bar charts
plt.show()
```
